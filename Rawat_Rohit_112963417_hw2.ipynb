````python
import imutils #To avoid cropping
def drawMatches(img1, kp1, img2, kp2, matches):
    """
    My own implementation of cv2.drawMatches as OpenCV 2.4.9
    does not have this function available but it's supported in
    OpenCV 3.0.0

    This function takes in two images with their associated 
    keypoints, as well as a list of DMatch data structure (matches) 
    that contains which keypoints matched in which images.

    An image will be produced where a montage is shown with
    the first image followed by the second image beside it.

    Keypoints are delineated with circles, while lines are connected
    between matching keypoints.

    img1,img2 - Grayscale images
    kp1,kp2 - Detected list of keypoints through any of the OpenCV keypoint 
              detection algorithms
    matches - A list of matches of corresponding keypoints through any
              OpenCV keypoint matching algorithm
    """

    # Create a new output image that concatenates the two images together
    # (a.k.a) a montage
    rows1 = img1.shape[0]
    cols1 = img1.shape[1]
    rows2 = img2.shape[0]
    cols2 = img2.shape[1]
    
    # Create the output image
    # The rows of the output are the largest between the two images
    # and the columns are simply the sum of the two together
    # The intent is to make this a colour image, so make this 3 channels
    out = np.zeros((max([rows1,rows2]),cols1+cols2,3), dtype='uint8')

    # Place the first image to the left
    out[:rows1,:cols1] = np.dstack([img1, img1, img1])

    # Place the next image to the right of it
    out[:rows2,cols1:] = np.dstack([img2, img2, img2])

    # For each pair of points we have between both images
    # draw circles, then connect a line between them
    for mat in matches:

        # Get the matching keypoints for each of the images
        img1_idx = mat[0].queryIdx
        img2_idx = mat[0].trainIdx

        # x - columns
        # y - rows
        (x1,y1) = kp1[img1_idx].pt
        (x2,y2) = kp2[img2_idx].pt

        # Draw a small circle at both co-ordinates
        # radius 4
        # colour blue
        # thickness = 1
        cv2.circle(out, (int(x1),int(y1)), 4, (255, 0, 0), 1)   
        cv2.circle(out, (int(x2)+cols1,int(y2)), 4, (255, 0, 0), 1)

        # Draw a line in between the two points
        # thickness = 1
        # colour blue
        cv2.line(out, (int(x1),int(y1)), (int(x2)+cols1,int(y2)), (0,255,0), 2)

    # Also return the image if you'd like a copy
    return out

# Read image
img_input = cv2.imread('SourceImages/sift_input.JPG', 0)

# initiate SIFT detector
siftobj = cv2.xfeatures2d.SIFT_create()
# find the keypoints and descriptors with SIFT
kp1, desc1 = siftobj.detectAndCompute(img_input,None)
# Darw keypoints on the image
res1 = cv2.drawKeypoints (img_input, kp1, None)
# ===== This is your first output =====


# rotate image
rows,cols = img_input.shape
#rotm = cv2.getRotationMatrix2D((cols/2,rows/2),300,1)
#rot_image = cv2.warpAffine(img_input,rotm,(cols,rows))
rot_image = imutils.rotate_bound(img_input, 60) #To avoid cropping after rotation
# find the keypoints and descriptors on the rotated image
kp2, desc2 = siftobj.detectAndCompute(rot_image,None)
# Darw keypoints on the rotated image
# ===== This is your second output =====
res2 = cv2.drawKeypoints (rot_image, kp2, None)

# ====== Plot functions, DO NOT CHANGE =====
# Plot result images
plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
plt.imshow(res1, 'gray')
plt.title('original img')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(res2, 'gray')
plt.title('rotated img')
plt.axis('off')
# ==========================================

# compute feature matching
bfm = cv2.BFMatcher()
ms = bfm.knnMatch(desc1, desc2, k = 2)
# Apply ratio test
good_matches = [] # Append filtered matches to this list
for m,n in ms:
    if m.distance < 0.1*n.distance:
        good_matches.append([m])        
# draw matching results with the given drawMatches function
res3 = drawMatches(img_input, kp1, rot_image, kp2, good_matches)

# ===== This is your third output =====
#print(good_matches)

# ====== Plot functions, DO NOT CHANGE =====
plt.figure(figsize=(12,8))
plt.imshow(res3)
plt.title('matching')
plt.axis('off')
# ==========================================

#print(np.array(ms))
#print(ms[0])

# estimate similarity transform
if len(good_matches) > 4:
    # Extract location of good matches
    #p1 = np.zeros((len(good_matches), 2), dtype=np.float32)
    #p2 = np.zeros((len(good_matches), 2), dtype=np.float32)
    p1 = np.float32([kp1[m[0].queryIdx].pt for m in good_matches ]).reshape(-1,1,2)
    p2 = np.float32([ kp2[m[0].trainIdx].pt for m in good_matches ]).reshape(-1,1,2)
    #for i, m in enumerate(good_matches):
     # p1[i, :] = kp1[good_matches[0].queryIdx].pt
     # p2[i, :] = kp2[[0].trainIdx].pt
    

    # find perspective transform matrix using RANSAC
    rot, mask = cv2.findHomography(p2, p1, cv2.RANSAC)
    print ("Transformation Matrix = \n", rot)


    # mapping rotataed image back with the calculated rotation matrix
    rows2, cols2 = rot_image.shape
    res4 = cv2.warpPerspective(rot_image, rot, (cols, rows))
    #rows2, cols2 = res41.shape
    #rotm2 = cv2.getRotationMatrix2D((cols2/2,rows2/2),300,1)
    #res4 = cv2.warpAffine(res41,rotm2,(cols2,rows2))
    #res4 = cv2.warpPerspective(res42, rot, (cols2, rows2))

    # ===== This is your fourth output =====
else:
    print ("Not enough matches are found - %d/%d" % (len(good_matches),4))

# ====== Plot functions, DO NOT CHANGE =====
# plot result images
plt.figure(figsize=(12,8))
plt.subplot(1, 2, 1)
plt.imshow(img_input, 'gray')
plt.title('original img')
plt.axis('off')
    
plt.subplot(1, 2, 2)
plt.imshow(res4, 'gray')
plt.title('recovered img')  
plt.axis('off')
# ==========================================
````

````python
from skimage.feature import hog
# Compute HOG features for the images
def computeHOGfeatures(img):
  
  desc, himg = hog(img, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1,1), visualize=True, multichannel=True)  
  return [desc, himg]  
  # Implement your computeHOGfeatures() here

# Compute HOG descriptors
descriptors = []
for i in range(0, 20):
  res = computeHOGfeatures(image_list[i][0])
  descriptors.append(res)
# ===== Display your second graph here =====
#print(np.array(descriptors).shape)
#print(descriptors[0])

plt.figure(figsize=(18,15))
plt.imshow(descriptors[0][1])
for i in range(20):
  plt.subplot(2, 10, i+1)
  plt.imshow(np.array(descriptors[i][1]))
  plt.title(np.array(image_list[i][1]))
  plt.axis('off')
#plt.imshow(descriptors[5][1])

# Split the data and labels into train and test set

train_list = []
train_ll = []
test_list = []
test_ll = []
for i in range(0,16):
  train_list.append(descriptors[i][0])
  train_ll.append(image_list[i][1]) 
  
for i in range(16,20):
  test_list.append(descriptors[i][0])
  test_ll.append(image_list[i][1]) 
````
